```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline

target_matrix = np.array([[1, 2],
                          [2, 4],
                          [3, 6],
                          [4, 8]])
df_target = pd.DataFrame(target_matrix)
df_target.columns = ["col1", "col2"]
print("* df_target :", "\n")
print(df_target, "\n")

corr_matrix = df_target.corr()
print("* correlation matrix of df_target :", "\n")
print(corr_matrix, "\n")

rank_of_target_matrix = np.linalg.matrix_rank(target_matrix)
print("* rank of target_matrix :", rank_of_target_matrix)

x = target_matrix[:, 0]
y = target_matrix[:, 1]

plt.figure(figsize = (12.5, 7.5))

plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("scatter plot of df_target")

plt.tight_layout()
plt.show()
```
