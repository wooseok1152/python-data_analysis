```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline

print("!!! sample of correlation matrix and rank in 2D !!!", "\n")

target_matrix = np.array([[1, 2],
                          [2, 4],
                          [3, 6],
                          [4, 8]])
df_target = pd.DataFrame(target_matrix)
df_target.columns = ["col1", "col2"]
print("* df_target :", "\n")
print(df_target, "\n")

corr_matrix = df_target.corr()
print("* correlation matrix of df_target :", "\n")
print(corr_matrix, "\n")

rank_of_target_matrix = np.linalg.matrix_rank(target_matrix)
print("* rank of target_matrix :", rank_of_target_matrix)

x = target_matrix[:, 0]
y = target_matrix[:, 1]

plt.figure(figsize = (12.5, 7.5))

plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("scatter plot of df_target")

plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
 

print("!!! sample of correlation matrix and rank in 3D !!!", "\n")

target_matrix = np.array([[10, -5,  20],
                          [20, -10, 40],
                          [30, -15, 60],
                          [40, -20, 80],
                          [50, -25, 100],
                          [60, -30, 120],
                          [70, -35, 140],
                          [80, -40, 160],
                          [90, -45, 180],
                          [100, -50, 200]])
df_target = pd.DataFrame(target_matrix)
df_target.columns = ["col1", "col2", "col3"]
print("* df_target :", "\n")
print(df_target, "\n")

corr_matrix = df_target.corr()
print("* correlation matrix of df_target :", "\n")
print(corr_matrix, "\n")

rank_of_target_matrix = np.linalg.matrix_rank(target_matrix)
print("* rank of target_matrix :", rank_of_target_matrix)

x = target_matrix[:, 0]
y = target_matrix[:, 1]
z = target_matrix[:, 2]
 
fig = plt.figure(figsize = (12.5, 7.5))

ax = plt.axes(projection ="3d")
ax.scatter3D(x, y, z)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.title("scatter plot of df_target")

ax.view_init(20,70)
plt.tight_layout()
plt.show()

```
