```python
# 데이터에 오류가 있는 상태로 집계를 실시(상품의 개수는 26개인데, 해당 pivot table의 column의 개수가 99개로 늘어나 있는 것을 확인할 수 있음)

uriage_purchase_dataset["purchase_date"] = pd.to_datetime(uriage_purchase_dataset["purchase_date"])
uriage_purchase_dataset["purchase_month"] = uriage_purchase_dataset["purchase_date"].dt.strftime("%Y%m")
pivot_table = uriage_purchase_dataset.pivot_table(index="purchase_month", columns="item_name", values="item_price", aggfunc="sum")
pivot_table
```
