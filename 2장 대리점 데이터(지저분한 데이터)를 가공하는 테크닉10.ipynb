# 테크닉 15 : 금액의 결측치를 수정하자

```python
print("* uriage_purchase_dataset 내 결측치 포함 컬럼 확인 :")
display(uriage_purchase_dataset.isnull().any()) # pd.Dataframe.isnull().any() : 특정 컬럼에 null값이 하나라도 포함되어 있으면 해당 컬럼에 대해선 'True'가, 
                                                #                               null값이 하나라도 포함되어 있지 않으면 'False'가 표기됨
print("* item_price 컬럼에 null값이 있음을 확인할 수 있음")                                       
```

```python
"""

- sql 문법을 사용하여 결측치 대체

"""

uriage_purchase_dataset = sqldf("""
                                    SELECT purchase_date,
                                            item_name,
                                            CASE 
                                                WHEN item_price IS NULL THEN MAX(item_price) OVER(PARTITION BY item_name)
                                                ELSE item_price
                                            END AS item_price,
                                            customer_name,
                                            purchase_month
                                    FROM uriage_purchase_dataset
                                """)
display(uriage_purchase_dataset)                                
```

```python
"""

- pandas 문법을 사용하여 결측치 대체

"""

flg_for_is_null = uriage_purchase_dataset["item_price"].isnull()

# boolean indexing 실시
for item in list(uriage_purchase_dataset.loc[flg_for_is_null, "item_name"].unique()):
    
    # 'item_name'이 'item'에 저장된 값에 해당하고 'item_price'값이 null이 아닌 행들의 'item_price'값을 요소로 가지는 Series를 추출한 후, 해당 Series에서 가장 최대값을 가져와 'price'변수에 저장   
    price = uriage_purchase_dataset.loc[(~flg_for_is_null) & (uriage_purchase_dataset["item_name"] == item), "item_price"].max()
    
    # 'item_name'이 'item'에 저장된 값에 해당하고 'item_price'가 null인 행들의 'item_price'값을 'price'에 저장된 값으로 변경
    uriage_purchase_dataset.loc[(flg_for_is_null) & (uriage_purchase_dataset["item_name"] == item), "item_price"] = price

display(uriage_purchase_dataset)
```

```python
print("* item_price 컬럼에 결측치 수정이 제대로 되었는지 확인 :")
display(uriage_purchase_dataset.isnull().any())
```

```python
print("* kokyaku_customer_dataset 내 등록일 값이 '숫자 형식'인 데이터를 조회한 후 'date 형식'으로 변경 :")

data_having_date_as_integer = sqldf("""
                                          SELECT 고객이름,
                                                 지역,
                                                 date(datetime('1900-01-01', '+' || 등록일 || ' days')) AS 등록일
                                          FROM   kokyaku_customer_dataset 
                                          WHERE  등록일 REGEXP('^[0-9]+$')
                                   """)
display(data_having_date_as_integer)

print("* kokyaku_customer_dataset 내 등록일 값이 'datetime 형식'인 데이터를 조회한 후 'date 형식'으로 변경 :")

data_having_date_as_datetime = sqldf("""
                                          SELECT 고객이름,
                                                 지역,
                                                 date(등록일) AS 등록일
                                          FROM   kokyaku_customer_dataset 
                                          WHERE  등록일 NOT REGEXP('^[0-9]+$')
                                   """)
display(data_having_date_as_datetime)      

print("* 위에서 산출한 두 데이터프레임 합치기 :")
kokyaku_customer_dataset = pd.concat([data_having_date_as_integer, data_having_date_as_datetime])
kokyaku_customer_dataset = kokyaku_customer_dataset.astype({"등록일" : "datetime64[ns]"})
display(kokyaku_customer_dataset)

print("* dtypes of kokyaku_customer_dataset :")
print(kokyaku_customer_dataset.dtypes)
```

```python
print("* count_by_month_and_item :")
count_by_month_and_item = sqldf("""
                                    SELECT purchase_month,
                                           item_name,
                                           COUNT(1) AS CNT
                                    FROM   import_dataset
                                    WHERE  purchase_month IS NOT NULL and item_name IS NOT NULL
                                    GROUP BY purchase_month, item_name
                                """)
display(count_by_month_and_item)                                

print("* pivot table of count_by_month_and_item :")
pivot_table = count_by_month_and_item.pivot_table(index = "purchase_month", columns = "item_name", fill_value = 0)
display(pivot_table)
```

```python
print("* price_by_month_and_item :")
price_by_month_and_item = sqldf("""
                                    SELECT purchase_month,
                                           item_name,
                                           SUM(item_price) AS price
                                    FROM   import_dataset
                                    WHERE  purchase_month IS NOT NULL and item_name IS NOT NULL
                                    GROUP BY purchase_month, item_name
                                """)
display(price_by_month_and_item)                                

print("* pivot table of price_by_month_and_item :")
pivot_table = price_by_month_and_item.pivot_table(index = "purchase_month", columns = "item_name", fill_value = 0)
display(pivot_table)
```

```python
print("* count_by_month_and_customer :")
count_by_month_and_customer = sqldf("""
                                    SELECT purchase_month,
                                           고객이름,
                                           COUNT(1) AS CNT
                                    FROM   import_dataset
                                    GROUP BY purchase_month, 고객이름
                                """)
display(count_by_month_and_customer)                                

print("* pivot table of count_by_month_and_customer :")
pivot_table = count_by_month_and_customer.pivot_table(index = "purchase_month", columns = "고객이름", fill_value = 0)
display(pivot_table)
```
