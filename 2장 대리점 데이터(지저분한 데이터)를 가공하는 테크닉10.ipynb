```python
# 데이터에 오류가 있는 상태로 집계를 실시(상품의 개수는 26개인데, 해당 pivot table의 column의 개수가 99개로 늘어나 있는 것을 확인할 수 있음)
uriage_purchase_dataset["purchase_date"] = pd.to_datetime(uriage_purchase_dataset["purchase_date"])
uriage_purchase_dataset["purchase_month"] = uriage_purchase_dataset["purchase_date"].dt.strftime("%Y%m")
uriage_purchase_dataset
pivot_table = uriage_purchase_dataset.pivot_table(index="purchase_month", columns="item_name", values="item_price", aggfunc="sum")
pivot_table
```

```python
# 상품명의 유니크 수 확인
print("unique한 상품명의 개수 :",len(list(pd.unique(uriage_purchase_dataset["item_name"]))), "\n")
print(list(pd.unique(uriage_purchase_dataset["item_name"])))
```

```python
# 상품명의 오류 수정(상품명 내 알파벳을 전부 대문자로 전환, 상품명에 포함된 띄어쓰기를 삭제)
uriage_purchase_dataset["item_name"] = uriage_purchase_dataset["item_name"].str.upper()
uriage_purchase_dataset["item_name"] = uriage_purchase_dataset["item_name"].str.replace(" ", "")
uriage_purchase_dataset["item_name"]
```

```python
# 상품명 컬럼의 오류가 잘 수정되었는지 최종 확인
len(pd.unique(uriage_purchase_dataset["item_name"]))
```

```python
# 특정 데이터 값이 null인지 아닌지 파악
uriage_purchase_dataset.isnull()
```

```python
# item_price 컬럼에 null값이 있음을 확인
# pd.Dataframe.isnull().any() : 특정 컬럼에 null값이 하나라도 포함되어 있으면 해당 컬럼에 대해선 'True'가, null값이 하나라도 포함되어 있지 않으면 'False'가 표기됨
uriage_purchase_dataset.isnull().any()
```

```python
flg_for_is_null = uriage_purchase_dataset["item_price"].isnull()

# boolean indexing 실시
for item in list(uriage_purchase_dataset.loc[flg_for_is_null, "item_name"].unique()):
    
    # 'item_name'이 'item'에 저장된 값에 해당하고 'item_price'값이 null이 아닌 행들의 'item_price'값을 요소로 가지는 Series를 추출한 후, 해당 Series에서 가장 최대값을 가져와 'price'변수에 저장   
    price = uriage_purchase_dataset.loc[(~flg_for_is_null) & (uriage_purchase_dataset["item_name"] == item), "item_price"].max()
    
    # 'item_name'이 'item'에 저장된 값에 해당하고 'item_price'가 null인 행들의 'item_price'값을 'price'에 저장된 값으로 변경
    uriage_purchase_dataset.loc[(flg_for_is_null) & (uriage_purchase_dataset["item_name"] == item), "item_price"] = price

uriage_purchase_dataset
```

```python
'''boolean을 요소로 가진 Series 변수 앞에 '~'을 붙이면, True가 False로 False가 True로 바뀐 Series가 반환된다.'''
~flg_for_is_null
```

```python
# item_price 결측치 수정이 제대로 되었는지 최종 확인1
uriage_purchase_dataset.isnull().any()
```

```python
# item_price 결측치 수정이 제대로 되었는지 최종 확인2
for item in sorted(list(uriage_purchase_dataset["item_name"].unique())):
    
    print(item + "의 최고가 : " + str(uriage_purchase_dataset.loc[uriage_purchase_dataset["item_name"] == item, ["item_price"]].max().values[0]) + " " + item + "의 최저가 : " + str(uriage_purchase_dataset.loc[uriage_purchase_dataset["item_name"] == item, ["item_price"]].min().values[0]))
```

```python
# 고객정보 데이터셋의 고객이름 오류 수정
kokyaku_customer_dataset["고객이름"] = kokyaku_customer_dataset["고객이름"].str.replace(" ", "")
kokyaku_customer_dataset["고객이름"] = kokyaku_customer_dataset["고객이름"].str.replace("  ", "")

kokyaku_customer_dataset
```

```python
# kokyaku_customer_dataset 내 컬럼들의 데이터 유형 확인 
kokyaku_customer_dataset.dtypes
```

```python
# 고객정보 데이터셋 내 등록일 컬럼내 날짜 형식이 아닌 숫자 형식으로 적혀진 데이터 값들을 알아내는 방법
flg_is_just_digit = kokyaku_customer_dataset["등록일"].astype("str").str.isdigit()
print(flg_is_just_digit, "\n")
print("등록일 컬럼에서 단지 숫자로 표기된 데이터 값의 총 개수 :", flg_is_just_digit.sum())
```

```python
# pd.to_timedelta() : 등록일 컬럼 내 숫자로 표기된 값들을 '시간' 형식으로 변환하는 메소드
result_error_fixed = pd.to_timedelta(kokyaku_customer_dataset.loc[flg_is_just_digit, "등록일"].astype("float"), unit="D") + pd.to_datetime("1900/01/01")
display(result_error_fixed)
```

```python
# 등록일 컬럼 내 날짜 형식으로 표기된 값들을 요소로 가진 Serise의 domain을 datetime으로 변환하기(위 Series의 domain과 통일시키기)
correct_Series = pd.to_datetime(kokyaku_customer_dataset.loc[~flg_is_just_digit, "등록일"])
display(correct_Series)
```

```python
# 위의 두 결과를 union하고, 해당 결과를 고객정보 데이터 셋으로 갱신
kokyaku_customer_dataset["등록일"] = pd.concat([result_error_fixed, correct_Series])
display(kokyaku_customer_dataset)
print(kokyaku_customer_dataset.dtypes)
```

```python
# 등록일 컬럼의 모든 데이터 값들이 datetime domain으로 잘 변환되었는지 검토
kokyaku_customer_dataset["등록일"].astype("str").str.isdigit().sum()
```

```python
# 등록일 컬럼으로부터 등록월을 추출하기
kokyaku_customer_dataset["등록연월"] = kokyaku_customer_dataset["등록일"].dt.strftime("%Y%m")
display(kokyaku_customer_dataset["등록연월"])

result = kokyaku_customer_dataset.groupby("등록연월").count()["고객이름"]
result
```
